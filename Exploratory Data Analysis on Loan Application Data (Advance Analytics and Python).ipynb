```python
categorical_columns = filtered_dataframe2.select_dtypes(include='object')
```

```python
categorical_columns.columns
```

```python
categorical_columns = categorical_columns.drop(columns = ['NAME_CONTRACT_STATUS'])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
l1 = categorical_columns.columns
```

```python
for i in range(len(l1)):
    categorical_columns[l1[i]] = le.fit_transform(categorical_columns[l1[i]])
```

```python
filtered_dataframe2 = filtered_dataframe2.drop(columns=['NAME_CONTRACT_TYPE_x', 'CODE_GENDER', 'FLAG_OWN_CAR',
       'FLAG_OWN_REALTY', 'NAME_TYPE_SUITE', 'NAME_INCOME_TYPE',
       'NAME_EDUCATION_TYPE', 'NAME_FAMILY_STATUS', 'NAME_HOUSING_TYPE',
       'OCCUPATION_TYPE', 'ORGANIZATION_TYPE', 'NAME_CONTRACT_TYPE_y',
       'NAME_CASH_LOAN_PURPOSE', 'CODE_REJECT_REASON', 'NAME_CLIENT_TYPE'])
```

```python
filtered_dataframe2 = pd.concat([filtered_dataframe2,categorical_columns],axis = 1)
```

```python
filtered_dataframe2 = filtered_dataframe2.dropna(axis = 0,how='any')
```

```python
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(filtered_dataframe2.drop(columns=['NAME_CONTRACT_STATUS','AGE','YEARS_OF_EXP'],axis = 1)
                                                 ,filtered_dataframe2['NAME_CONTRACT_STATUS'],test_size = 0.25)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import accuracy_score
```

```python
mod = LogisticRegression().fit(x_train, y_train)
```

```python
y_pred = mod.predict(x_test)
```

```python
acc = accuracy_score(y_test,y_pred)
acc
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```

```python
cnf_matrix = confusion_matrix(y_test, y_pred)
cnf_matrix
```
